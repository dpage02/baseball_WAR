```python
# depedencies
import pandas as pd
from sqlalchemy import create_engine

# from password import password 
```

```python
# enter postgress password
password = 'password'
```

```python
# create engine
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/etlproject_db')
```

```python
# create connecteion 
connection = engine.connect()
```

```python
# download csv file 
csv_file = "Resources/FanGraphs Leaderboard (3).csv"
```

```python
# turn csv 
df_file = pd.read_csv(csv_file)
```

```python
df_file.head()
```

```python
# creating column headers 
df_file.columns = ['Name','playerid','Team','Pos','PA','IP','Primary WAR', 'Total WAR']
df_file.head()
```

```python
# spliting df into position players and pitchers
position_df = df_file.loc[df_file['Pos'] != 'P']
pitcher_df = df_file.loc[df_file['Pos'] == 'P']
position_df.head()
```

```python
#dropping IP for position players
# clean_pos_df = position_df.drop('IP',axis=1, inplace=True)
clean_pos_df = position_df.drop(['IP'],axis=1)
print(clean_pos_df)

# dropping PA for pitchers
clean_pitch_df = pitcher_df.drop(['PA'], axis=1)
print(clean_pitch_df)
```

```python
# sending dataframes to postgress
clean_pos_df.to_sql('position_players',con=engine,index=False, if_exists='replace')
```

```python
clean_pitch_df.to_sql('pitchers',con=engine,index=False, if_exists='replace')
```
